```python
import re
import matplotlib.pyplot as plt
from youtube_transcript_api import YouTubeTranscriptApi
from collections import Counter
from wordcloud import WordCloud
```

In the baove code snippet, we are importing the essesntial library for the transript extraction from the downloded package.

```python
# Function to get the transcript of a YouTube video
def get_transcript(video_id):
    transcript_list = YouTubeTranscriptApi.get_transcript(video_id)
    transcript = " ".join([item['text'] for item in transcript_list])
    return transcript
```

```python
# Function to clean and normalize the transcript
def clean_text(text):
    # Convert text to lowercase
    text = text.lower()
    # Remove punctuation and special characters
    text = re.sub(r'[^\w\s]', '', text)
    return text
```

```python
# Function to check the regular expression patterns
def check_patterns(text):
    patterns = {
        "alphabetic_strings": r'^[A-Za-z]+$',
        "lowercase_alphabetic_ending_b": r'^[a-z]*b$',
        "each_a_preceded_followed_by_b": r'^b(ab)*b$',
        "consecutive_repeated_words": r'\b(\w+)\s+\1\b',
        "start_with_integer_end_with_word": r'^\d+.*\b\w+\b$',
        "contain_grotto_and_raven": r'(?=.*\bgrotto\b)(?=.*\braven\b)',
        "first_word_of_sentence": r'^\b\w+\b'
    }
    results = {key: bool(re.search(pattern, text)) for key, pattern in patterns.items()}
    return results
```

```python
# Function to plot word frequency distribution
def plot_word_frequency(word_frequencies):
    common_words = word_frequencies.most_common(10)
    words, frequencies = zip(*common_words)

    plt.figure(figsize=(10, 6))
    plt.bar(words, frequencies)
    plt.xlabel('Words')
    plt.ylabel('Frequencies')
    plt.title('Top 10 Most Frequent Words')
    plt.show()
```

```python
# Function to plot word cloud
def plot_word_cloud(word_frequencies):
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_frequencies)

    plt.figure(figsize=(10, 8))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title('Word Cloud')
    plt.show()
```

```python
# Function to plot word length distribution
def plot_word_length_distribution(words):
    word_lengths = [len(word) for word in words]

    plt.figure(figsize=(12, 8))
    plt.hist(word_lengths, bins=range(1, max(word_lengths)+1), edgecolor='black')
    plt.xlabel('Word Length')
    plt.ylabel('Frequency')
    plt.title('Word Length Distribution')
    plt.show()
```

```python
# Main
video_id = 'fLeJJPxua3E'  # Replace with your YouTube video ID
transcript = get_transcript(video_id)
cleaned_transcript = clean_text(transcript)

print("Cleaned Transcript:", cleaned_transcript)
```

```python
words = cleaned_transcript.split()
total_words = len(words)
unique_words = len(set(words))

print("Total number of words:", total_words)
print("Total number of unique words:", unique_words)
```

```python
word_frequencies = Counter(words)
print("**************")
print("Word frequencies:", word_frequencies)

most_frequent_word = word_frequencies.most_common(1)[0]
least_frequent_word = word_frequencies.most_common()[-1]
print("**************")

print("Most frequent word:", most_frequent_word)
print("**************")

print("Least frequent word:", least_frequent_word)

longest_word = max(words, key=len)
print("**************")

print("Longest word:", longest_word)
print("**************")
```
